```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
from sodapy import Socrata

import warnings
warnings.filterwarnings('ignore')
```

```python
# edge coordinates of map
    # ie: bottom left corner of the map is -97.82, 30.218
bounds = [-97.82, -97.68, 30.218,30.325]

# map source = openstreetmap.org
map_img = plt.imread("atx_map.png")
plt.figure(figsize = (20,10))
plt.imshow(map_img, aspect="equal")
plt.axis('off')
plt.show()
```

```python
# get data from live API data. updated weekly
def make_request(where_date = None):
    if where_date is None:
        result = Socrata("data.austintexas.gov", None).get("fdj4-gpfu",
                                                           order="occ_date DESC",
                                                           limit=2000)
    else:
        where_clause = "occ_date_time < '{}'".format(where_date)
        result = Socrata("data.austintexas.gov", None).get("fdj4-gpfu",
                                                           where=where_clause,
                                                           order="occ_date_time DESC",
                                                           limit=2000)
    return result
```

```python
#we can only get 2,000 records at once. this isn't really a lot when I want to limit to certain map dimensions
#so I'm going to make a few requests, and then combine all the data. should give me at least 10,000 rows to start with
all_data = []
date = None
for i in range(5):
    result = make_request(where_date=date)
    result_df = pd.DataFrame(result)
    all_data.append(result_df)
    date = min((all_data[len(all_data)-1]).occ_date_time)
```

```python
#we got our 10,000 rows!
raw_data = pd.concat(all_data, sort=False)
raw_data.shape
```

```python
raw_data.head()
```

```python
limited_data = raw_data.copy()
limited_data.drop(columns = ['clearance_status','clearance_date'], axis=1, inplace=True)
limited_data['latitude'] = pd.to_numeric(limited_data.latitude , downcast='float')
limited_data['longitude'] = pd.to_numeric(limited_data.longitude , downcast='float')
limited_data['occ_date'] = pd.to_datetime(limited_data.occ_date_time)
```

```python
limited_data = limited_data[((limited_data.longitude >= -97.82) & (limited_data.longitude <= -97.68))]
limited_data = limited_data[((limited_data.latitude >= 30.218) & (limited_data.latitude <= 30.325))]
limited_data.shape
#so we have 3,710 data points that are on our map
```

```python
#reorganize new limited data
limited_data = limited_data.sort_values(by='occ_date_time').reset_index(drop=True)
min(limited_data['longitude'])
```

```python
fig, ax = plt.subplots(figsize = (8,7))

ax.scatter(limited_data.longitude, limited_data.latitude, zorder=1, alpha= 0.2, c='r', s=10)

ax.set_title('austin crime locales')
ax.set_xlim(bounds[0],bounds[1])
ax.set_ylim(bounds[2],bounds[3])
ax.imshow(map_img, zorder=0, extent = bounds, aspect= 'equal')
```
